```python
from sklearn.linear_model import LinearRegression

# Veri: metrekare -> fiyat (bin TL)
X = [[50], [60], [70], [80], [90]]
y = [150, 180, 210, 240, 270]

model = LinearRegression()
model.fit(X, y)

print("Eğim (metrekare başına artış):", model.coef_[0])
print("Sabit (başlangıç):", model.intercept_)

tahmin_100 = model.predict([[100]])[0]
print("100 m² tahmini fiyat:", tahmin_100, "bin TL")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

X = [
    [50, 2],
    [60, 2],
    [70, 3],
    [80, 3],
    [90, 4],
    [100, 4],
    [110, 4],
    [120, 5]
]
y = [150, 180, 220, 250, 300, 340, 380, 420]

def run_split(train_size):
    X_train, y_train = X[:train_size], y[:train_size]
    X_test, y_test = X[train_size:], y[train_size:]

    model = LinearRegression()
    model.fit(X_train, y_train)

    train_mae = mean_absolute_error(y_train, model.predict(X_train))
    test_mae = mean_absolute_error(y_test, model.predict(X_test))

    print(f"--- Train size: {train_size} / {len(X)} ---")
    print("Train MAE:", train_mae)
    print("Test  MAE:", test_mae)
    print("Test tahmin:", model.predict(X_test))
    print("Gerçek     :", y_test)
    print()

# %75 eğitim (%25 test) => 6 train, 2 test
run_split(6)

# %50 eğitim (%50 test) => 4 train, 4 test
run_split(4)
```

```python
X = [
    [50, 2],
    [60, 2],
    [70, 3],
    [80, 3],
    [90, 4],
    [100, 4],
    [110, 4],
    [120, 5]
]
y = [150, 180, 220, 250, 300, 340, 380, 420]

X_train = X[:6]
y_train = y[:6]
X_test = X[6:]
y_test = y[6:]

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

tree = DecisionTreeRegressor(max_depth=2, random_state=42)
tree.fit(X_train, y_train)

train_mae = mean_absolute_error(y_train, tree.predict(X_train))
test_mae = mean_absolute_error(y_test, tree.predict(X_test))

print("TREE Train MAE:", train_mae)
print("TREE Test  MAE:", test_mae)
print("TREE Test tahmin:", tree.predict(X_test))
print("Gerçek:", y_test)

```

```python
# Orijinal veri: [m2, oda, yas]
X_raw = [
    [50, 2, 20],
    [60, 2, 18],
    [70, 3, 15],
    [80, 3, 12],
    [90, 4, 8],
    [100, 4, 5],
    [110, 4, 3],
    [120, 5, 1]
]
y = [150, 180, 220, 250, 300, 340, 380, 420]

def age_penalty(yas):
    if yas < 5:
        return 0
    elif yas < 10:
        return -10
    else:
        return -20

# Yeni X: [m2, oda, yas_cezasi]
X = [[m2, oda, age_penalty(yas)] for m2, oda, yas in X_raw]

print("Yeni X örnekleri (m2, oda, ceza):")
for row in X:
    print(row)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

X_train, y_train = X[:6], y[:6]
X_test, y_test = X[6:], y[6:]

model = LinearRegression()
model.fit(X_train, y_train)

train_mae = mean_absolute_error(y_train, model.predict(X_train))
test_mae = mean_absolute_error(y_test, model.predict(X_test))

print("Train MAE:", train_mae)
print("Test  MAE:", test_mae)
print("Test tahmin:", model.predict(X_test))
print("Gerçek:", y_test)

print("Katsayılar [m2, oda, ceza]:", model.coef_)
print("Sabit:", model.intercept_)

```
